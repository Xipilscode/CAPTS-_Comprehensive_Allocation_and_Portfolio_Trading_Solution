```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
slices = [25,35,45] # weights of asset list, input weights from monte carlo data, using for loop TBD.
assets = ['Crypto', 'Stocks', 'Commodities'] # asset list
cols = ['c','b','g'] # color list in in order of above assets
```

```python
plt.pie(slices,
        labels = assets,
        colors = cols,
        startangle = 90,
        shadow = False,
        explode =(0,0,0.1),
        autopct = '%1.1f%%'
) # sets up the pie chart with descriptions

plt.title('Suggested Stakeholder Investemnts by Weights')
plt.show()
```
